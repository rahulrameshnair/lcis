```python
import networkx as nx
from networkx import Graph
DG = nx.DiGraph()
DG.add_node("meow", label='cat', node_color='green')
DG.add_edge("meow", "yawn", label='sound', edge_color='red')
DG.add_edge("snore", "meow", label='sound', edge_color='red')
print(DG)
nx.draw(DG,  with_labels=True, node_color='blue', edge_color='red')
```
