```python
import networkx as nx
from networkx import Graph
DG = nx.DiGraph()
li = [1,2,3,4,5]
li2=['a']
for i in li:
    
    for j in li2:
        DG.add_node(i)
        DG.add_edge(j,i)
        DG.add_edge(i,"cat")
nx.draw(DG,  with_labels=True, node_color='blue', edge_color='red')
```
